# Central Limit Theorem

The Central Limit Theorem (CLT) is a fundamental concept in statistics and probability theory. It states that the sum or average of a large number of independent and identically distributed (i.i.d.) random variables will have an approximately normal distribution, regardless of the shape of the original distribution.

In simpler terms, the CLT tells us that when we take a large enough sample from any population, the distribution of the sample means or sums will tend to follow a bell-shaped curve, known as a normal distribution, even if the individual data points themselves do not follow a normal distribution. This holds true as long as the sample size is sufficiently large.

The implications of the Central Limit Theorem are profound. It allows us to make inferences about population parameters based on sample statistics. It is the foundation for many statistical techniques, such as hypothesis testing, confidence intervals, and regression analysis, which rely on the assumption of normality.

## An example: heights of people in a population

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the distribution, heights in cm
min_height = 150
max_height = 190
n_people = 100_000

# Generate random heights
heights = np.random.randint(min_height, max_height, n_people)

# Plot the distribution
plt.hist(heights, bins=40, density=True, histtype="step", label="Population")
plt.title("Height Distribution of Population and Sample Means")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
# plt.show()

# Sampling
sample_size = 20
n_samples = 500

# Initialize the array of sample means
sample_means = []
for _ in range(n_samples):
    sample = np.random.choice(heights, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.hist(sample_means, bins=40, density=True, histtype="step", label="Sample Means")
plt.legend()
plt.show()
```
