# Finding π

Finding the value of PI using the Monte Carlo method

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from tqdm.notebook import tqdm

# number of points
n = 1000

# generate two sets of random points between 0 and 1.
x = np.random.rand(n)
y = np.random.rand(n)

# calculate the distance from the origin
d = np.sqrt(x**2 + y**2)

# Here we count the number of points that are inside the circle.
# Intuitively, we can think of the circle as a dart board and the points as darts.
# If the distance from the origin is less than 1, then the point is inside the circle.
count = 0
pis = []
for i in tqdm(range(n)):
    if d[i] < 1:
        count += 1

    pi = 4 * count / (i + 1)
    pis.append(pi)


# Plotting the values of PI
plt.plot(pis)
plt.title("Finding $π$")
plt.xlabel("Iteration")
plt.ylabel("PI")
plt.axhline(y=np.pi, color="r", linestyle="--")
plt.legend(["simulated $π$", "real $π$"])
plt.plot(len(pis) - 1, pis[-1], "bo")
plt.text(len(pis) - 60, pis[-1] + 0.05, f"{pis[-1]:.4f}")
plt.show()
```

Alternatively, speaking of darts, we can visualize the points and the circle.

```python
plt.figure(figsize=(5, 5))
for i in range(n):
    # think of x[i] and y[i] as the coordinates of the dart. and calculate the distance from 0.5, 0.5
    dist = np.sqrt((x[i] - 0.5) ** 2 + (y[i] - 0.5) ** 2)
    if dist < 0.5:
        plt.plot(x[i], y[i], "o", markersize=3, color="purple")
    else:
        plt.plot(x[i], y[i], "o", markersize=3, color="pink")

plt.title("Finding $π$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
